```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
optimisation_results = pd.read_json("experiment_cosmeticsshop/optimisation_results.json")
```

```python
final_results = pd.read_json("experiment_cosmeticsshop/results.json")
```

```python
optimisation_results
```

```python
final_results.sort_values('ndcgk_10', ascending=False)
```

```python
params_df = pd.concat([optimisation_results, pd.DataFrame(optimisation_results['params'].tolist())], axis=1)
params_df['algorithm'] = params_df.identifier.apply(lambda x: x.split('(')[0])
```

```python
params_df
```

```python
fig = plt.figure(figsize=(15, 10))
gs = plt.GridSpec(4, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, 0:1])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])
ax4 = fig.add_subplot(gs[2, 0:1])
ax5 = fig.add_subplot(gs[3, 0:1])

seaborn.pointplot(
    data = params_df[params_df.algorithm=='TARSItemKNNLee'], 
    ax=ax1, 
    x='W', 
    y='ndcgk_10',
    hue='similarity'
)

seaborn.pointplot(
    data = params_df[params_df.algorithm=='TARSItemKNNLiu'], 
    ax=ax2, 
    x='fit_decay', 
    y='ndcgk_10',
    capsize=0.1,
)

seaborn.pointplot(
    data = params_df[params_df.algorithm=='TARSItemKNNLiu'], 
    ax=ax3, 
    x='predict_decay', 
    y='ndcgk_10',
    capsize=0.1,
)

seaborn.pointplot(
    data = params_df[params_df.algorithm=='TARSItemKNNDing'], 
    ax=ax4, 
    x='predict_decay', 
    y='ndcgk_10',
    hue='similarity',
)

seaborn.pointplot(
    data = params_df[params_df.algorithm=='TARSItemKNNLiu2012'], 
    ax=ax5, 
    x='decay', 
    y='ndcgk_10',
)


plt.tight_layout()
```
